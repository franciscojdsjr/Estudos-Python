```python
import pandas as pd
import numpy as np
```

```python
dados_vendas_clientes = pd.read_json('dados_vendas_clientes.json')

dados_vendas = pd.json_normalize(dados_vendas_clientes['dados_vendas'])

dados_vendas
```

```python
colunas_vendas = list(dados_vendas.columns)

dados_vendas = dados_vendas.explode(colunas_vendas[1:]).reset_index(drop = True)

dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip()).astype(np.float64)

dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_locacao = pd.read_json('dados_locacao_imoveis.json')
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])

dados_locacao
```

```python
colunas_vendas = list(dados_locacao.columns)

dados_locacao = dados_locacao.explode(colunas_vendas[1:]).reset_index(drop = True)

dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$','').replace('reais','').replace(',','.').strip()).astype(np.float64)

```
