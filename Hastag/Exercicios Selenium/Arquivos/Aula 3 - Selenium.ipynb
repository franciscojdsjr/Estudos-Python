```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os


servico = Service(ChromeDriverManager().install())
#navegador = webdriver.Chrome(service=servico)
```

```python
caminho = os.getcwd() #pega o caminho da pasta do arquivo raiz
arquivo =  caminho + r"\Pagina Hashtag.html"

navegador = webdriver.Chrome(service=servico)
navegador.get(arquivo)
navegador.maximize_window()

navegador.find_element(By.ID, 'fullname').send_keys('Francisco')
navegador.find_element(By.ID, 'email').send_keys('franciscojunior@gmail.com')
navegador.find_element(By.ID, '_form_176_submit').click()

navegador.minimize_window()

navegador.quit()
```
